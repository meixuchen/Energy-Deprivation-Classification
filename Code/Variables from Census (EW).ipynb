```python
import pandas as pd
import glob
import seaborn as sns
import numpy as np
```

```python
#census files
path = glob.glob('2021 census/*.csv')
files = [pd.read_csv(f, header = 5, index_col = 1).iloc[0:35673] for f in path]
```

```python
files[0]['lone parent with dependent childern'] = files[0]['%.4']
files[0] = files[0].iloc[:,-1:]
```

```python
files[1]['Aged 75 and above'] = files[1]['Aged 75 to 79 years'] + files[1]['Aged 80 to 84 years'] + files[1]['Aged 85 years and over']
files[1] = files[1].iloc[:,[2,-1]]
```

```python
files[2]['large household size >5 people'] = files[2]['%.1'] + files[2]['%.2'] + files[2]['%.3']
files[2] = files[2].iloc[:,[-1]]
```

```python
files[3]['unpaid care 20h+'] = files[3]['Provides 20 to 49 hours unpaid care a week'] + files[3]['Provides 50 or more hours unpaid care a week']
files[3] = files[3].iloc[:,[-1]]
```

```python
files[4] = files[4].iloc[:,2:]
files[4].columns = ['Detached','Semi-detached','Terraced','Flat','Shared house']
```

```python
files[5] = files[5][['No central heating','Renewable energy only','Electric only']]
```

```python
files[6]['under occupancy'] = files[6].iloc[:,-2:].sum(axis = 1)
files[6] = files[6].iloc[:,[-1]]
```

```python
# Tenure
files[7] = files[7].iloc[:,[3,6,7,9]]
```

```python
files[8] = files[8].iloc[:,[-1]]
files[8].columns = ['Elementary occupations']
```

```python
files[9]['Unemployed'] = files[9].iloc[:,[5,9]].sum(axis= 1)
files[9]['Economically active: Employee: Part-time'] = files[9].iloc[:,[2,3,4,6,7,8]].sum(axis = 1)
files[9] = files[9].iloc[:,10:]
```

```python
files[10]['ethnic_minority'] = files[10].iloc[:,1]-files[10].iloc[:,2]
files[10] = files[10].iloc[:,[-1]]
```

```python
files[11]['English not as main language'] = files[11].iloc[:,[-1]]
files[11] = files[11][['English not as main language']]
```

```python
files[12]['Entry-level education and below'] = files[12]['No qualifications'] + files[12]['Level 1 and entry level qualifications'] 
files[12] = files[12][['Entry-level education and below']]
```

```python
census_total = pd.concat(files, axis = 1).iloc[1:35673]
census_total
```

```python
census_total.to_csv('Vars_census_EW.csv')
```
